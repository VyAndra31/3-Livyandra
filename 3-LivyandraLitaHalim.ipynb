### Test case 1 : write a function to convert Celsius to Fahrenheit

***You can only write 1 line in a function***

Formula: `F = (C * 9/5) + 32`

```python
#write and continue the code down here
def celcius_to_fahrenheit(suhu):
    return (suhu * 9/5) + 32

#Test
print(celcius_to_fahrenheit(0))    # Expected: 32.0
print(celcius_to_fahrenheit(100))  # Expected: 212.0
```

### Test case 2 : write a function to count vowels in a string

Count vowels: a, i, u, e, o (case-insensitive).

```python
def vowels_count(text):
    count = 0
    text = text.lower()
    vowels = "aiueo"
    for huruf in text:
      if huruf in vowels:
        count += 1
    return count

# Test
print(vowels_count("Hello World"))  # Expected: 3  (e,o,o)
print(vowels_count("PYTHON"))       # Expected: 1  (o)
```

# Medium level
### Test case 3 : write a function to summarize shopping cart

Given a list of prices, return (total, average, max). Use built-in functions when possible.

```python
def summary(prices):
    total = sum(prices)
    average = total / len(prices)
    max_prices = max(prices)
    return f"({total}, {average}, {max_prices})"

# Test
print(summary([10, 20, 30]))         # Expected: (60, 20.0, 30)
print(summary([5]))                  # Expected: (5, 5.0, 5)
```

# Hard level
### Test case 4 : write a function to remove duplicates while keeping order

Example: `[1,2,1,3,2] -> [1,2,3]`

```python
def remove_duplicates(items):
    result = []
    for n in items:
      if n not in result:
        result.append(n)
    return result

# Test
print(remove_duplicates([1, 2, 1, 3, 2]))                 # Expected: [1, 2, 3]
print(remove_duplicates(["a", "b", "a", "c", "b"]))       # Expected: ['a', 'b', 'c']
```

### Test case 5 : create a class StudentGradeBook

Requirements:
- `__init__(self, name)` sets name and empty list of scores
- `add_score(self, score)` adds score (0-100). If invalid, return "Invalid score"
- `average(self)` returns average score (float). If no scores, return 0
- `status(self)` returns:
  - "Pass" if average >= 75
  - "Remedial" if 60 <= average < 75
  - "Fail" if average < 60


```python
class StudentGradeBook:
    def __init__(self, name):
        self.list_score = []
        self.nama = name

    def add_score(self, nilai):
        if nilai < 0 or nilai > 100:
          return "Invalid Score"
        else:
          self.list_score.append(nilai)
          return f"Added {nilai}"


    def average(self):
        total = sum(self.list_score)
        if len(self.list_score) == 0:
          return 0
        return total / len(self.list_score)

    def status(self):
        ratarata = self.average()
        if ratarata >= 75:
          return "Pass"
        elif ratarata >= 60 and ratarata < 75:
          return "Remedial"
        else:
          return "Fail"

# Test
s = StudentGradeBook("Alya")
print(s.average())              # Expected: 0
print(s.status())               # Expected: Fail
print(s.add_score(80))          # Expected: Added 80
print(s.add_score(70))          # Expected: Added 70
print(s.add_score(120))         # Expected: Invalid score
print(s.average())              # Expected: 75.0
print(s.status())               # Expected: Pass
```
