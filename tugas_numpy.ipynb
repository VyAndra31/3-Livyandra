## **📌 Soal 4: Boolean Masking & `np.where()`**

Diberikan array suhu berikut:


```python
temps = np.array([29, 31, 35, 28, 33, 27, 30, 36, 32, 26])
print(temps)
```

Kerjakan hal berikut:
1. Ambil semua suhu yang **lebih dari 30**
2. Hitung berapa banyak suhu yang **lebih dari atau sama dengan 32**
3. Buat array label baru bernama `status` dengan aturan:
   - `"panas"` jika suhu > 32
   - `"normal"` jika suhu 29 s.d. 32
   - `"dingin"` jika suhu < 29

**Hint:** untuk poin (3), kamu bisa menggunakan `np.where()` bertingkat.


```python
# suhu lebih dari 30
temps[temps > 30]
```

```python
# banyak suhu lebih dari sama dengan 32
np.sum(temps >= 32)
```

```python
# array label baru bernama status
status = np.where(temps > 32, 'panas', np.where(temps >= 29, 'normal', 'dingin'))
print(status)
```

## **📌 Soal 5: View vs Copy**

Buat array berikut:


```python
x = np.array([10, 20, 30, 40, 50, 60])
print(x)
```

Lalu lakukan:
1. Buat `view_part` dari elemen indeks 1 sampai 4 menggunakan slicing
2. Buat `copy_part` dari elemen indeks `[1, 2, 3, 4]` menggunakan advanced indexing
3. Ubah elemen pertama `view_part` menjadi `999`
4. Ubah elemen pertama `copy_part` menjadi `-100`
5. Tampilkan `x`, `view_part`, dan `copy_part`

Terakhir, tuliskan dengan singkat:
- mana yang menyebabkan array asli berubah?
- mana yang tidak?


```python
# Tulis jawabanmu di sini
view_part = x[1:5]
copy_part = x[[1,2,3,4]]
view_part[0] = 999
copy_part[0] = -100
print(x)
print(view_part)
print(copy_part)

# Tambahkan penjelasan singkat dalam komentar juga.
# nah kalau view_part ini dia pake basic indexing dimana dia cuma nyalin reference nya gitu
# makanya dia tetep keubah tuh x nya yang index ke 1 nya juga ikutan jadi 999
# kalau copy_part dia advance indexing dimana dia langsung copy dan bikin array baru
# jadi ga akan ngaruh ke array x nya
```

## **📌 Soal 6: Statistik Dasar & Missing Value (`NaN`)**

Gunakan array berikut:


```python
values = np.array([12, 15, np.nan, 18, 21, np.nan, 24, 30])
print(values)
```

Tampilkan:
1. Rata-rata biasa dengan `np.mean()`
2. Rata-rata yang mengabaikan `NaN` dengan `np.nanmean()`
3. Standar deviasi yang mengabaikan `NaN` dengan `np.nanstd()`
4. Nilai minimum dan maksimum yang mengabaikan `NaN`

Lalu jelaskan singkat kenapa hasil `np.mean()` bisa berbeda dari `np.nanmean()`.


```python
# rata rata biasa
np.mean(values)
```

```python
# rata rata yang mengabaikan NaN
np.nanmean(values)
```

```python
# standar deviasi yang mengabaikan NaN
np.nanstd(values)
```

```python
# nilai minimum dan maximum yang mengabaikan NaN
print(np.nanmin(values))
print(np.nanmax(values))
```

```python
# kalau np.mean() biasa, dia menghitung rata rata dari semua nilai yang ada dalam array
# kalau ada NaN nya, maka hasilnya akan menjadi NaN karena nilai NaN nya juga ikut kehitung
# kalau np.nanmean() dia mengabaikan data NaN jadi yang dihitung hanya rata rata dari data yang valid saja
```

## **📌 Soal 7: Broadcasting untuk Normalisasi**

Diberikan matriks fitur berikut:


```python
X = np.array([
    [170, 65, 28],
    [165, 55, 24],
    [180, 75, 30],
    [175, 70, 29]
], dtype=float)

print(X)
```

Anggap setiap kolom merepresentasikan fitur yang berbeda.  
Lakukan normalisasi z-score sederhana dengan rumus:

$$
Z = \frac{X - \text{mean kolom}}{\text{std kolom}}
$$

Tampilkan:
1. Mean per kolom
2. Std per kolom
3. Hasil normalisasi `Z`

**Hint:** gunakan `axis=0` dan manfaatkan broadcasting NumPy.


```python
# mean per kolom
np.mean(X, axis=0)
```

```python
# standar deviasi per kolom
np.std(X, axis=0)
```

```python
# hasil normalisasi Z
z = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
print(z)
```

## **📌 Soal 8: Reshape & Mini Challenge**

Gunakan array 1 dimensi berikut:


```python
z = np.arange(1, 25)
print(z)
```

Kerjakan hal berikut:
1. Ubah `z` menjadi matriks berukuran `6x4`
2. Ubah hasilnya menjadi bentuk `3x2x4`
3. Ambil kembali seluruh data tersebut menjadi array 1 dimensi
4. Bonus: buat array window 4 elemen yang bergeser dari `z`, sehingga bentuk akhirnya menjadi:

`[[1,2,3,4], [2,3,4,5], ..., [21,22,23,24]]`

**Hint bonus:** boleh gunakan loop **atau** `sliding_window_view` jika kamu sudah tahu.


```python
# ubah z menjadi matriks 6x4
z_6x4 = z.reshape(6,4)
print(z_6x4)
```

```python
# ubah hasilnya jadi bentuk 3x2x4
z_3x2x4 = z_6x4.reshape(3,2,4)
print(z_3x2x4)
```

```python
# buat lagi jadi array 1 dimensi
print(z_3x2x4.flatten())
```

```python
# buat array window 4 elemen yang bergeser dari z
from numpy.lib.stride_tricks import sliding_window_view
windows = sliding_window_view(z, 4)
print(windows)
```
